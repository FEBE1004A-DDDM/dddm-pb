## Line Plot

To visualize the data we:
* Load data from a table
    * `Table().read_table(filename)``

```python
ratios = Table.read_table('ratios.csv')
ratios
```

Then we lot a line plot using:
* `tbl.plot(x_column, y_column)`
* `tbl.plot(x_column)`

```python
ratios.plot('Age', 'F:M Ratio')
```

* Print the name of the plot

```python
ratios.plot('Age', 'F:M Ratio')
print('Age vs F:M Ratio')  # <--- Print the title of the plot
```

Add a title to the plot using:
* `plots.title('string')`

```python
ratios.plot('Age', 'F:M Ratio')
plots.title('Age vs F:M Ratio')
```

In case you need to adjust the limits of the x-axis use:
* `plots.xlim(start, end)`
  
In case you need to adjust the limits of the y-axis use:
* `plots.ylim(start, end)`

```python
ratios.plot('Age', 'F:M Ratio')
plots.title('Age vs F:M Ratio');
plots.ylim(0, 100);
```
